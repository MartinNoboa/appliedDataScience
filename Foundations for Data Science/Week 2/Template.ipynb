```python
import scipy.stats as st

# Given parameters
mean = 502
std_dev = 115

# Values for the range
a = 400
b = 590

# Calculate Z-scores
z_a = (a - mean) / std_dev
z_b = (b - mean) / std_dev

# Find probabilities using CDF
probability_a = st.norm.cdf(z_a)
probability_b = st.norm.cdf(z_b)

# Calculate the percentage
percentage = (probability_b - probability_a) * 100

print("Percentage of students scoring between 400 and 590:", round(percentage, 2), "%")
```
