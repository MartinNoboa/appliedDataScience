```python
# Given values
P = 11000  # initial investment
r = 0.15   # annual growth rate (as a decimal)
n = 12     # number of times interest is compounded per year
t = 1      # number of years
PMT = 8000 # monthly contribution

# Calculate the future value
FV = P * (1 + r/n)**(n*t) + PMT * (((1 + r/n)**(n*t) - 1) / (r/n))

# Print the result
print("Total value after one year:", round(FV, 2))
```
